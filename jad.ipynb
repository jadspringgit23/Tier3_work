**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
import pandas as pd
import matplotlib.pyplot as plt

# URL to the Excel file
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# Read the Excel file into a DataFrame
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col=None)

# Melt the DataFrame
melted_properties = pd.melt(properties, id_vars=['Unnamed: 0'], var_name='London Borough', value_name='Average Price')

# Clean up column names
melted_properties.columns = ['ID', 'London Borough', 'Average Price']

# Convert 'Average Price' to numeric, replacing non-numeric values with NaN
melted_properties['Average Price'] = pd.to_numeric(melted_properties['Average Price'], errors='coerce')

# Filter for entries related to 'City of London'
city_of_london_entries = melted_properties[melted_properties['London Borough'] == 'City of London']

# Display the first few rows of the subset DataFrame
print(city_of_london_entries.head())

# Get the number of entries for 'City of London'
num_entries_city_of_london = len(city_of_london_entries)
print(f"\nNumber of entries for 'City of London': {num_entries_city_of_london}")

# Line plot of Month against Average Price for 'City of London'
plt.figure(figsize=(12, 6))
plt.plot(city_of_london_entries['ID'], city_of_london_entries['Average Price'], marker='o', linestyle='-')
plt.title('Average House Price in City of London Over Time')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
import pandas as pd

def create_price_ratio(data):
    """
    Calculate the ratio of house prices in 2018 to the prices in 1998 for each London borough.

    Parameters:
    - data: DataFrame containing London house prices data.

    Returns:
    - ratio_df: DataFrame containing London boroughs and their corresponding price ratios.
    """
    # Melt the DataFrame
    melted_data = pd.melt(data, id_vars=['Unnamed: 0'], var_name='London Borough', value_name='Average Price')

    # Clean up column names
    melted_data.columns = ['Unnamed: 0', 'London Borough', 'Average Price']

    # Extract the year from the 'Date' column
    melted_data['Year'] = melted_data['Unnamed: 0'].dt.year

    # Filter data for the years 1998 and 2018
    filtered_data = melted_data[(melted_data['Year'] == 1998) | (melted_data['Year'] == 2018)]

    # Pivot the data to have 1998 and 2018 prices as columns
    pivoted_data = filtered_data.pivot_table(index='London Borough', columns='Year', values='Average Price',aggfunc='mean')

    # Calculate the price ratio for 2018 to 1998
    pivoted_data['Price Ratio'] = pivoted_data[2018] / pivoted_data[1998]

    # Reset the index to have the borough as a regular column
    ratio_df = pivoted_data.reset_index()

    return ratio_df[['London Borough', 'Price Ratio']]

# URL to the Excel file
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# Read the Excel file into a DataFrame
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col=None)

# Use the function to calculate the price ratio
price_ratio_df = create_price_ratio(properties)

# Display the resulting DataFrame
print(price_ratio_df)
```
